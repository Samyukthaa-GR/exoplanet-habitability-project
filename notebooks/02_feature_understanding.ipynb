```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/physics_dataset_v2.csv")
```

```python
df.shape
df.isna().sum()
```

```python
plt.figure()
plt.hist(df['teq_recalc'], bins=50)
plt.xlabel("Equilibrium Temperature (K)")
plt.ylabel("Count")
plt.title("Distribution of Equilibrium Temperature")
plt.show()
```

```python
df['teq_recalc'].skew()
```

```python
df['log_teq'] = np.log10(df['teq_recalc'])

plt.figure()
plt.hist(df['log_teq'], bins=50)
plt.xlabel("log10(Equilibrium Temperature)")
plt.ylabel("Count")
plt.title("Log-Transformed Temperature Distribution")
plt.show()
```

```python
df['log_teq'].skew()
```

```python
plt.figure()
plt.hist(df['teq_recalc'], bins=50)
plt.title("Raw Teq")
plt.show()

plt.figure()
plt.hist(df['log_teq'], bins=50)
plt.title("Log Teq")
plt.show()
```

```python
df['teq_recalc'].quantile([0.01, 0.05, 0.95, 0.99])
```

```python
print("Raw Teq quantiles:")
print(df['teq_recalc'].quantile([0.01, 0.05, 0.95, 0.99]))

print("\nLog Teq quantiles:")
print(df['log_teq'].quantile([0.01, 0.05, 0.95, 0.99]))
```
